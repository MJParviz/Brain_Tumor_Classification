## Introduction to Convolutional Neural Networks (CNNs)

This repository presents a comprehensive implementation of Convolutional Neural Networks (CNNs) tailored for image classification, with a specific focus on MRI brain images. CNNs are a sophisticated class of deep learning models renowned for their efficacy in processing visual data, particularly in tasks such as image recognition and object detection.

### Project Overview

The primary objective of this project is to design, train, and evaluate a CNN model utilizing PyTorch, aimed at classifying MRI images into four distinct categories:

-   No Tumor
-   Glioma
-   Meningioma
-   Pituitary Tumor

The dataset is meticulously curated to ensure balanced representation across classes, with each image resized to 512x512 pixels and normalized to enhance training performance.

### Key Components

-   **Dataset Preparation**: A custom dataset class is implemented to facilitate the loading and preprocessing of images.
-   **Model Architecture**: The architecture comprises multiple convolutional and pooling layers for feature extraction, followed by fully connected layers for classification.
-   **Training Process**: The model is trained using appropriate loss functions and optimizers to ensure effective learning.
-   **Evaluation Metrics**: Performance is assessed through metrics such as accuracy, precision, recall, and confusion matrices.
-   **Regularization Techniques**: Strategies such as dropout and batch normalization are employed to improve model robustness and mitigate overfitting.

### Load Data

```python
import os
import pandas as pd

dataset_dir = "/content/drive/MyDrive/Tumor/Data"
data = []

for tumor_type in os.listdir(dataset_dir):
    tumor_type_path = os.path.join(dataset_dir, tumor_type)

    for filename in os.listdir(tumor_type_path):
        img_path = os.path.join(tumor_type_path, filename)

        label = tumor_type

        data.append((img_path, label))

df = pd.DataFrame(data, columns=['image_path', 'label'])
df
```

### Visualizing Data and Checking Data Balance

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,6))
colors = ['#4285f4', '#ea4335', '#fbbc05', '#34a853']
plt.rcParams.update({'font.size': 14})
plt.pie([len([x for x in df.label if x=='pituitary']),
         len([x for x in df.label if x=='notumor']),
         len([x for x in df.label if x=='meningioma']),
         len([x for x in df.label if x=='glioma'])],
        labels=['pituitary','notumor', 'meningioma', 'glioma'],
        colors=colors, autopct='%.1f%%', explode=(0.025,0.025,0.025,0.025),
        startangle=30);
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

value_counts_train = train['label'].value_counts()
value_counts_train_df = value_counts_train.reset_index()
value_counts_train_df.columns = ['Category', 'Count']
sns.barplot(x='Category', y='Count', data=value_counts_train_df, ax=axes[0])
axes[0].set_title('Value Counts of Categories in Train')
plt.xticks(rotation=45)

value_counts_test = test['label'].value_counts()
value_counts_test_df = value_counts_test.reset_index()
value_counts_test_df.columns = ['Category', 'Count']
sns.barplot(x='Category', y='Count', data=value_counts_test_df, ax=axes[1])
axes[1].set_title('Value Counts of Categories in Test')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
test_data, valid_data = train_test_split(test, test_size=0.5, random_state=42, shuffle = True , stratify=test['label'])

print(f"Training set shape: {train.shape}")
print(f"Validation set shape: {valid_data.shape}")
print(f"Testing set shape: {test_data.shape}")
```

### Undersampling the Data for Equal Class Counts

```python
from imblearn.under_sampling import RandomUnderSampler

X_train = train.drop(columns=['label'])
y_train = train['label']

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

train_data = pd.concat([X_resampled, y_resampled], axis=1)

train_data['label'].value_counts()
```

### Using ImageDataGenerator for Data Augmentation

```python
import warnings
warnings.filterwarnings("ignore")
from tensorflow.keras.preprocessing.image import ImageDataGenerator

tr_gen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=10,
                                   brightness_range=(0.85, 1.15),
                                   width_shift_range=0.002,
                                   height_shift_range=0.002,
                                   shear_range=12.5,
                                   zoom_range=0,
                                   horizontal_flip=True,
                                   vertical_flip=False,
                                   fill_mode="nearest"
                                   )
ts_gen = ImageDataGenerator(rescale=1./255)

input_size = (150, 150)
batch_size = 32

train_generator = tr_gen.flow_from_dataframe(
    dataframe=train,
    x_col='image_path',
    y_col='label',
    target_size=input_size,
    shuffle = True,
    batch_size=batch_size,
    class_mode='categorical',
    seed=111
)

valid_generator = ts_gen.flow_from_dataframe(
    dataframe=valid_data,
    x_col='image_path',
    y_col='label',
    target_size=input_size,
    shuffle = True,
    batch_size=batch_size,
    class_mode='categorical',
    seed=111
)

test_generator = ts_gen.flow_from_dataframe(
    dataframe=test_data,
    x_col='image_path',
    y_col='label',
    target_size=input_size,
    shuffle=False,
    batch_size=batch_size,
    class_mode='categorical',
    seed=111
)
```

### Viewing Sample Images from the Dataset

```python
import numpy as np

sample_images, sample_labels = next(train_generator)

plt.figure(figsize=(12, 12))
for i in range(16):
    image = sample_images[i]
    label_index = np.argmax(sample_labels[i])
    label = list(train_generator.class_indices.keys())[label_index]

    plt.subplot(4, 4, i+1)
    plt.imshow(image)
    plt.title(label, color='k', fontsize=12)
    plt.axis("off")

plt.show()
```

### Modeling with Convolutional Neural Networks (CNNs)

```python
# Building Model
from keras.utils import plot_model
from tensorflow.keras import models
from tensorflow.keras import models
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Training Model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Define the model architecture
model = models.Sequential([

    # Convolutional layer 1
    Conv2D(32, (4, 4), activation="relu", input_shape=(150,150,3)),
    MaxPooling2D(pool_size=(3, 3)),

    # Convolutional layer 2
    Conv2D(64, (4, 4), activation="relu"),
    MaxPooling2D(pool_size=(3, 3)),

    # Convolutional layer 3
    Conv2D(128, (4, 4), activation="relu"),
    MaxPooling2D(pool_size=(3, 3)),

    # Convolutional layer 4
    Conv2D(128, (4, 4), activation="relu"),
    Flatten(),

    # Full connect layers
    Dense(512, activation="relu"),
    Dropout(0.35, seed=111),
    Dense(4, activation="softmax")
])

model.summary()

optimizer = Adam(learning_rate=0.001, beta_1=0.869, beta_2=0.995)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics= ['accuracy'])
```

```python
# Stop training if loss doesn't keep decreasing.
model_es = EarlyStopping(monitor='loss', min_delta=1e-9, patience=8, verbose=True)
model_rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=5, verbose=True)

# Training the model
history = model.fit(train_generator,
                    steps_per_epoch=40,
                    epochs=100,
                    validation_data=test_generator,
                    validation_steps=3,
                    callbacks=[model_es, model_rlr])
```

```python
  loss, accuracy = model.evaluate(test_generator, steps=test_generator.samples//batch_size)
print(f"Test Loss: {loss:0.5f}")
print(f"Test Accuracy: {accuracy:0.5f}")
```

### Plotting Loss and Accuracy of training and validation data

```python
train_acc = history.history['accuracy']
train_loss = history.history['loss']

val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']

index_loss = np.argmin(val_loss)
val_lowest = val_loss[index_loss]

index_acc = np.argmax(val_acc)
val_highest = val_acc[index_acc]

Epochs = [i+1 for i in range(len(train_acc))]

loss_label = f'Best epochs = {str(index_loss +1)}'
acc_label = f'Best epochs = {str(index_acc + 1)}'

#Training history

plt.figure(figsize= (20,8))
plt.style.use('fivethirtyeight')

plt.subplot(1,2,1)
plt.plot(Epochs , train_loss , 'r' , label = 'Training Loss')
plt.plot(Epochs , val_loss , 'g' , label = 'Validation Loss')
plt.scatter(index_loss + 1 , val_lowest , s = 150 , c = 'blue',label = loss_label)
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(Epochs , train_acc , 'r' , label = 'Training Accuracy')
plt.plot(Epochs , val_acc , 'g' , label = 'Validation Accuracy')
plt.scatter(index_acc + 1 , val_highest , s = 150 , c = 'blue',label = acc_label)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout
plt.show();
```

### Predict and Evaluating the Model


```python
preds = model.predict(test_generator)

y_pred = np.argmax(preds, axis=1)
```

```python
from sklearn.metrics import classification_report
classes = ['glioma', 'meningioma', 'notumor','pituitary']

print(classification_report(test_generator.classes, y_pred , target_names= classes ))
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(test_generator.classes, y_pred, average='weighted')
print("F1 Score: ", f1)
```

```python
train_score = model.evaluate(train_generator)
print(f"Training Loss: {train_score[0]}, Training Accuracy: {train_score[1]}")

valid_score = model.evaluate(valid_generator)
print(f"Validation Loss: {valid_score[0]}, Validation Accuracy: {valid_score[1]}")

test_score = model.evaluate(test_generator)
print(f"Test Loss: {test_score[0]}, Test Accuracy: {test_score[1]}")
```

### Plotting the Confusion Matrix


```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
import itertools

g_dict = test_generator.class_indices
classes = list(g_dict.keys())

# Confusion matrix
cm = confusion_matrix(test_generator.classes, y_pred)

print(cm)

plt.figure(figsize= (12, 8))
plt.imshow(cm, interpolation= 'nearest', cmap= plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation= 45)
plt.yticks(tick_marks, classes)


thresh = cm.max() / 2.
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, cm[i, j], horizontalalignment= 'center', color= 'white' if cm[i, j] > thresh else 'black')

plt.tight_layout()

plt.show()
```
